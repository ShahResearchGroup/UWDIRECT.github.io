# Expressiveness plot in L0

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_context('talk')
```

```python
df = pd.DataFrame(
    {
        'x1': np.random.normal(size=(400)),
        'x2': np.random.normal(size=(400)),
        'x3': np.random.normal(size=(400))
    }
)
```

```python
fig, ax = plt.subplots()
g = sns.histplot(df, x='x1', ax=ax)
ax.set_xlabel('Sum(X)')
```

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=500, n_features=2, centers=2)
```

```python
df = pd.DataFrame({'PC1': X[:,0], 'PC2': X[:,1], 'cluster': y})
```

```python
fig, ax = plt.subplots(figsize=(5,5))
sns.kdeplot(df, x='PC1', y='PC2', hue='cluster')
```
